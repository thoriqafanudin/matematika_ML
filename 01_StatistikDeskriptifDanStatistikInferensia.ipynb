```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
sns.set()
```

Menentukan rataan, grup 1 = tanpa pupuk, grup 2 = dengan pupuk

```python
rataan1 = 50
rataan2 = 51
```

Menentukan standar deviasi

```python
sd = 5
```

Ukuran sample dua kelompok

```python
sample1 = 50
sample2 = 40
```

Generate bilangan random

```python
test = np.random.randn(sample1)
test
```

```python
plt.plot(test)
```

```python
sns.distplot(test);
```

Generate data

```python
data1 = np.random.randn(sample1)*sd + rataan1
data2 = np.random.randn(sample2)*sd + rataan2
```

Menggabungkan sample size (jumlah data)

```python
kelompok_data = [sample1, sample2]
batas = [np.min(np.hstack((data1, data2))), np.max(np.hstack((data1, data2)))]
```

```python
batas
```

Plot distribusi kedua kelompok (statistik deskriptif)

```python
plt.figure(figsize=(7, 5))
a = sns.distplot(data1, hist=False, label='Data1')
a = sns.distplot(data2, hist=False, label='Data2')

a.set(xlabel='Data', ylabel='Density')
plt.legend();
```

Melakukan uji statistik (statistik inferential)

```python
fig, ax = plt.subplots(1,2, figsize=(9,5), dpi=100)

ax[0].violinplot(data1)
ax[0].plot(1+np.random.randn(sample1)/10, data1, 'o', color='green')
ax[0].set_ylim(batas)

ax[1].violinplot(data2)
ax[1].plot(1+np.random.randn(sample2)/10, data2, 'o', color='red')
ax[1].set_ylim(batas)

# Melakukan uji t-test
t, p = stats.ttest_ind(data1, data2)

# Mencetak hasil t-test di judul
sigtxt = ('', ' TIDAK')
plt.title(f'Dua kelompok{sigtxt[int(p>.05)]} berbeda secara signifikan! t({sum(kelompok_data)-2}) = {np.round(t,2)}, p = {np.round(p,6)}')

plt.show()
```
